```python
from selenium import webdriver

driver = webdriver.Chrome('/Users/apple1/Downloads/chromedriver 3')#chrome web driver ë²„ì „ì— ë§ê²Œ ê¹”ê¸°
driver.implicitly_wait(3)#ì›¹ ìì› ë¡œë“œë¥¼ ìœ„í•´ 3ì´ˆ ê¸°ë‹¤ë ¤ì¤Œ
driver.get('https://soundcloud.com/uiceheidd/righteous')#url ì ‘ê·¼
```

```python
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time 

body = driver.find_element_by_tag_name('body')

#ì¸ê¸°ìˆœ/ì‘ì„±ìˆœ ì„ íƒí•  ìˆ˜ ìˆëŠ” ì˜ì—­ í´ë¦­
#driver.find_element_by_xpath('//paper-button[@class="dropdown-trigger style-scope yt-dropdown-menu"]').click()
#ì¸ê¸°ìˆœ ì¹´í…Œê³ ë¦¬ í´ë¦­
#driver.find_element_by_xpath('//paper-listbox[@class="dropdown-content style-scope yt-dropdown-menu"]/a[1]').click()

num_pagedown = 100 #në²ˆ ë°‘ìœ¼ë¡œ ë‚´ë¦¬ëŠ” ê²ƒ
while num_pagedown:
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(2)
    num_pagedown -=1
    
page=driver.page_source
soup=BeautifulSoup(page,'html.parser')
comments=soup.find_all('yt-formatted-string',attrs={'class':'style-scope ytd-comment-renderer'})
```

```python
#content = [comments[n].string for n in range(0,len(comments))]

content=[]
for li in comments:
    comment=li.getText().split('\n')[0]
    if comment=='' : continue
    content.append(comment) #í•œ ì°¨ì›ìœ¼ë¡œ ë¬¶ìŒ 
    #content.append(li.getText().split('\n'))
print(content)
```

```python
#ì €ì¥í•˜ê¸°
with open('trigono C.txt', 'w',encoding='utf-8') as f:
    for i in content:
        f.write(i+"\n")
```
